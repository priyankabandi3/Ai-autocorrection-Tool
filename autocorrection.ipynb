```python
# ============================================
# AI AUTOCORRECT TOOL USING PYTHON
# ============================================

# STEP 1:
# Install required libraries before running:
#
# pip install textblob nltk pandas matplotlib
#
# If using Google Colab:
# !pip install textblob nltk pandas matplotlib

# ============================================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
import nltk

# Download NLTK data
nltk.download('punkt')

# ============================================
# SAMPLE DATASET
# ============================================

data = {
    'Wrong_Word': ['havv', 'pythn', 'dreem', 'goood', 'langauge'],
    'Correct_Word': ['have', 'python', 'dream', 'good', 'language']
}

df = pd.DataFrame(data)

print("\n===== DATASET =====")
print(df)

# ============================================
# USER INPUT
# ============================================

text = input("\nEnter a sentence with spelling mistakes:\n")

# ============================================
# AUTOCORRECTION USING TEXTBLOB
# ============================================

blob = TextBlob(text)

corrected_text = blob.correct()

# ============================================
# OUTPUT
# ============================================

print("\n===== RESULTS =====")
print("Original Text  :", text)
print("Corrected Text :", corrected_text)

# ============================================
# ANALYSIS
# ============================================

original_words = text.split()
corrected_words = str(corrected_text).split()

incorrect_count = 0

for o, c in zip(original_words, corrected_words):
    if o != c:
        incorrect_count += 1

correct_count = len(corrected_words) - incorrect_count

print("\n===== ANALYSIS =====")
print("Total Words      :", len(original_words))
print("Corrected Words  :", incorrect_count)
print("Unchanged Words  :", correct_count)

# ============================================
# ACCURACY CALCULATION
# ============================================

accuracy = (correct_count / len(original_words)) * 100

print("\nAccuracy: {:.2f}%".format(accuracy))

# ============================================
# VISUALIZATION
# ============================================

labels = ['Corrected Words', 'Unchanged Words']
values = [incorrect_count, correct_count]

plt.figure(figsize=(6,5))
plt.bar(labels, values)

plt.title("Autocorrect Word Analysis")
plt.xlabel("Category")
plt.ylabel("Count")

plt.show()

# ============================================
# END OF PROJECT
# ============================================
```
